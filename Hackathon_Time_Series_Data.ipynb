```python
import pandas as pd
```

```python
t1 = pd.read_csv('/content/EducationData20152016MetricOnly.csv')
t2 = pd.read_csv('/content/EducationData20172018MetricOnly.csv')
```

```python
time_series = pd.concat([t1,t2], ignore_index=True)
time_series
```

```python
times_series = time_series.drop(['leaid', 'city_location', 'agency_type', 'county_code', ], axis=1)
times_series
```

```python
time_series = time_series.dropna()

for col in time_series.columns:
  time_series = time_series[time_series[col] != 'Missing/not reported']

for col in time_series.columns:
  time_series = time_series[time_series[col] != 'Suppressed data']

for col in time_series.columns:
  time_series = time_series[time_series[col] != 'Not applicable']

time_series
```

```python
import numpy as np
time_series['District Student Performance Metric'] = 0.3 * np.array(time_series['read_test_pct_prof_midpt']).astype(np.float) + 0.3 * np.array(time_series['math_test_pct_prof_midpt']).astype(np.float) + 0.4 * np.array(time_series['grad_rate_midpt']).astype(np.float)
time_series
```

```python
time_series.to_excel('times_series_data.xlsx')  
```
